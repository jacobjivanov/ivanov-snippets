```python
import numpy as np
import matplotlib.pyplot as plt
from numba import njit

plt.rcParams['figure.dpi'] = 600

@njit
def oblique_shock(M1, theta_s, gamma=1.4):
    delta = np.arctan(
        2/np.tan(theta_s)
        * (M1**2*np.sin(theta_s)**2 - 1)
        / (M1**2*(gamma + np.cos(2*theta_s)) + 2)
    )

    num = (gamma-1)/2*M1**2*np.sin(theta_s)**2 + 1
    den = gamma*M1**2*np.sin(theta_s)**2 - (gamma-1)/2

    M2 = 1/np.sin(theta_s - delta)*np.sqrt(num/den)

    return delta, M2

@njit
def barV_to_barVcomps(barV, theta_s, delta):
    barVr = barV*np.cos(theta_s - delta)
    barVt = -barV*np.sin(theta_s - delta)

    W = np.empty(2)
    W[0] = barVr
    W[1] = barVt

    return W

@njit
def M_to_barV(M, gamma=1.4):
    return (2/((gamma-1)*M**2) + 1)**(-0.5)

@njit
def dWdtheta(theta, W, gamma):
    barVr = W[0]
    barVt = W[1]

    N = barVr*barVt**2 - (gamma-1)*(1 - barVr**2 - barVt**2)*(2*barVr + barVt/np.tan(theta))/2
    D = (gamma-1)*(1 - barVr**2 - barVt**2)/2 - barVt**2

    dW = np.empty(2)

    dW[0] = barVt
    dW[1] = N/D

    return dW

@njit
def rk45_step(theta, W, h, gamma):
    k1 = dWdtheta(theta, W, gamma)
    k2 = dWdtheta(theta + h*(1/5), W + h*(1/5)*k1, gamma)
    k3 = dWdtheta(theta + h*(3/10), W + h*(3/40)*k1 + h*(9/40)*k2, gamma)
    k4 = dWdtheta(theta + h*(4/5), W + h*(44/45)*k1 - h*(56/15)*k2 + h*(32/9)*k3, gamma)
    k5 = dWdtheta(theta + h*(8/9), W + h*(19372/6561)*k1 - h*(25360/2187)*k2 + h*(64448/6561)*k3 - h*(212/729)*k4, gamma)
    k6 = dWdtheta(theta + h, W + h*(9017/3168)*k1 - h*(355/33)*k2 + h*(46732/5247)*k3 + h*(49/176)*k4 - h*(5103/18656)*k5, gamma)

    # 5th order
    W5 = W + h*(35/384)*k1 + h*(500/1113)*k3 + h*(125/192)*k4 - h*(2187/6784)*k5 + h*(11/84)*k6

    # 4th order
    k7 = dWdtheta(theta + h, W5, gamma)

    W4 = W + h*(5179/57600)*k1 + h*(7571/16695)*k3 + h*(393/640)*k4 - h*(92097/339200)*k5 + h*(187/2100)*k6 + h*(1/40)*k7

    err = W5 - W4
    return W5, err


@njit
def find_theta_c(theta_s, M1, h = -1e-6, tol = 1e-9, gamma = 1.4):
    delta, M2 = oblique_shock(M1, theta_s, gamma)
    barV = M_to_barV(M2, gamma)
    W = barV_to_barVcomps(barV, theta_s, delta)
    theta = theta_s

    Vt_prev = W[1]

    while theta > 0:
        W_new, err = rk45_step(theta, W, h, gamma)
        err_norm = np.max(np.abs(err))

        if err_norm < tol:
            theta_new = theta + h
            Vt_new = W_new[1]

            # Cone detection: V_theta = 0
            if Vt_prev*Vt_new < 0:
                alpha = abs(Vt_prev)/(abs(Vt_prev)+abs(Vt_new))
                theta_c = theta + alpha*h

                return theta_c

            theta = theta_new
            W = W_new
            Vt_prev = Vt_new

            if err_norm == 0: s = 2.0
            else: s = 0.9*(tol/err_norm)**0.2

            h *= min(5.0, max(0.2, s))

        else:
            s = 0.9*(tol/err_norm)**0.25
            h *= max(0.1, s)
        
    return theta

@njit
def find_theta_c_full_solution(theta_s, M1, h=-1e-6, tol=1e-9, gamma=1.4):
    delta, M2 = oblique_shock(M1, theta_s, gamma)
    barV = M_to_barV(M2, gamma)
    W = barV_to_barVcomps(barV, theta_s, delta)

    theta = theta_s
    Vt_prev = W[1]

    # ---- Preallocate storage ----
    Nmax = 200
    theta_hist = np.empty(Nmax)
    Vr_hist = np.empty(Nmax)
    Vt_hist = np.empty(Nmax)

    i = 0
    theta_hist[i] = theta
    Vr_hist[i] = W[0]
    Vt_hist[i] = W[1]
    i += 1

    while theta > 0 and i < Nmax:
        W_new, err = rk45_step(theta, W, h, gamma)
        err_norm = np.max(np.abs(err))

        if err_norm < tol:
            theta_new = theta + h
            Vt_new = W_new[1]

            # ---- Store accepted step ----
            theta_hist[i] = theta_new
            Vr_hist[i] = W_new[0]
            Vt_hist[i] = W_new[1]
            i += 1

            # ---- Cone detection ----
            if Vt_prev * Vt_new < 0:
                alpha = abs(Vt_prev) / (abs(Vt_prev) + abs(Vt_new))
                theta_c = theta + alpha * h

                # Trim arrays to actual size
                return (
                    theta_c,
                    theta_hist[:i],
                    Vr_hist[:i],
                    Vt_hist[:i]
                )

            theta = theta_new
            W = W_new
            Vt_prev = Vt_new

            # ---- Step size control ----
            if err_norm == 0:
                s = 2.0
            else:
                s = 0.9 * (tol / err_norm)**0.2

            h *= min(5.0, max(0.2, s))

        else:
            s = 0.9 * (tol / err_norm)**0.25
            h *= max(0.1, s)

    # Fallback
    return theta, theta_hist[:i], Vr_hist[:i], Vt_hist[:i]
```

```python
gamma = 1.4
M1 = 2.0

mu = np.arcsin(1/M1)
theta_s = np.deg2rad(np.array([30.001, 30.05, 30.1, 31, 35, 45]))
color = ['red', 'orange', 'lightgreen', 'skyblue', 'royalblue', 'indigo']

plt.axvline(np.rad2deg(mu), linestyle = 'dashed', color = 'black')

for i in range(0, len(theta_s)):
    theta_c, theta, barVr, barVt = find_theta_c_full_solution(theta_s[i], M1, h = -1e-6, tol = 1e-9, gamma = gamma)

    N = barVr*barVt**2 - (gamma-1)*(1 - barVr**2 - barVt**2)*(2*barVr + barVt/np.tan(theta))/2
    D = (gamma-1)*(1 - barVr**2 - barVt**2)/2 - barVt**2

    plt.scatter(np.rad2deg(theta[-1]), D[-1], color = color[i], marker = 'x')
    plt.scatter(np.rad2deg(theta[0]), D[0], facecolors = 'none', edgecolor = color[i], marker = 'o')
    plt.semilogy(np.rad2deg(theta), D, label = r'$\theta_s =$' +' {0:.3f}º'.format(np.rad2deg(theta_s[i])), color = color[i])


plt.scatter(-1, 0, facecolors = 'none', edgecolor = 'black', marker = 'o', label = 'Shock (Start)')
plt.scatter(-1, 0, color = 'black', marker = 'x', label = 'Cone (End)')

plt.grid(); plt.legend(ncol = 2)
plt.gca().set(xlabel = r'Polar Angle, $\theta$ [º]', xlim = (0, 90), ylim = (1e-6, 1), ylabel = 'Second Derivative Term [-]', title = r'$D$ Term for Various Shock Angle, $M_1 = 2.0$, $\mu = $ 30º')
plt.show()
```

The reason why I was having a lot of difficulty solving near-minimum shock angles for a given Mach Number is that both the numerator and denominator for the second derivative term approach zero. This can be avoided by recognizing that the minimum shock angle is the Mach Wave Angle ($\mu$)!
